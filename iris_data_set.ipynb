```python
import pandas as pd
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_excel("train.xlsx")
```

```python
data.head()
```

```python
x=data.iloc[:,:-1]
```

```python
y=data.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x.values,y.values)
```

```python
predictions=model.predict([[7.9,3.8,6.4,2]])
```

```python
predictions
```

```python
data_test=pd.read_excel("test.xlsx")
```

```python
x_test=data_test.iloc[:,:-1]
```

```python
y_test=data_test.iloc[:,-1]
```

```python
x
```

```python
y_test
```

```python
x_test
```

```python
predictions=model.predict(x_test.values)
```

```python
predictions
```

```python
accuracy=y_test.values
```

```python
count =0
for i in range(len(accuracy)):
    if predictions[i]==accuracy[i]:
        count= count+1
```

```python
print((count*100)/len(accuracy))
```
